### 1. FakeNewsNet

We cannot download the entire dataset due to Twitter privacy policies and news publisher copy rights.  Social engagements and user information are not disclosed because of Twitter Policy.
However, we can download news articles from published websites and relevant social media data from Twitter.



#### 1.1 How to download the dataset

The scripts and procedures to download this dataset are clearly mentioned in their official github url as given below:

https://github.com/KaiDMML/FakeNewsNet



**Requirements:**

Data download scripts are writtern in python and requires python 3.6 + to run.


Twitter API keys are used for collecting data from Twitter. Try to create a developer account in Twitter account and make use of the following link to get Twitter API keys
https://developer.twitter.com/en/docs/basics/authentication/guides/access-tokens.html

Once you got the api keys which is an array of JSON object with attributes such as 

- app_key
- app_secret
- oauth_token
- oauth_token_secret 

you need to update the the corresponding values in the **tweet_keys_file.json** file located in **code/resources** folder which is available in the below fakenewsnet code script repo url

https://github.com/KaiDMML/FakeNewsNet

Then follow the steps as per the above url to download the data



#### 1.2 Folder Structure after download

Once you download the dataset,mainly two category folders are created as below

- politifact
- gossipcop



The downloaded dataset will have the following folder structure,





##### News Content

**news content.json:** This json includes all the meta information of the news articles collected using the provided news source URLs. This is a JSON object with attributes including:

- text is the text of the body of the news article.
- url is the news published url 
- images is a list of the URLs of all the images in the news article web page.
- publish date indicate the date that news article is published.
- source 
- authors

#### 2.1 How to download the dataset

You can download the dataset from the below URL:

https://drive.google.com/file/d/1IoTRrJNDJqvaG3hnUpnHQyGvPAJbO8y3/view

Description of the dataset is available from this URL;

https://www.uvic.ca/ecs/ece/isot/assets/docs/ISOT_Fake_News_Dataset_ReadMe.pdf


The dataset contains two types of articles fake and real News.

The dataset consists of two CSV files. The first file named “True.csv” contains more than 12,600
articles from reuter.com. The second file named “Fake.csv” contains more than 12,600 articles from
different fake news outlet resources. Each article contains the following information: article title, text,
type and the date the article was published on.
